```python
import pandas as pd

df_sales = pd.read_csv("pandas_practice.csv")
df_customers = pd.read_csv("customers.csv")
```

```python
df_sales["sales_per_degree"] = df_sales["sales_amount"] / df_sales["temperature"]
```

```python
df_sales["sales_per_degree"]
```

```python
df_sales["sales_per_degree"].mean()
```

```python
df_sales["sales_per_degree"] = df_sales["sales_per_degree"].fillna( 15.234313245468396 )
```

```python
df_sales["sales_per_degree"]
```

```python
df_sales["feels_like_weather"] = df_sales["temperature"].apply(
    lambda t: "Cold" if t < 15 else ("Warm" if t <= 25 else "Hot")
)
```
